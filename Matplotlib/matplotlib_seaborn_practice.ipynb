```python
%matplotlib inline
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib
import json

matplotlib.style.use('default') #Explain this
```

```python
period = 150
x = pd.period_range(datetime.now(), periods = period, freq='d')
x = x.to_timestamp().to_pydatetime()
y = np.random.randn(period,3).cumsum(0)

ts_line_chart = plt.plot(x,y)
```

```python
multi_line_chart = plt.plot(x,y[:,0], '+',
                            x,y[:,1], '-',
                            x,y[:,2], '*')
```

Figure Cheat sheet for Matplotlib: https://matplotlib.org/stable/_images/anatomy.png


```python
fig = plt.figure(figsize=(12,6))
## Main Plot
ax = fig.add_axes((0, 0, 1, 1)) ## 0,0 is coordination, 1,1 is the size
ax.set_title("Customize Axes Object")
ax.plot(x,y[:,1], color='b')
ax.set_xlabel('Date')
ax.set_ylabel('Cummulative Sum')

## Plot inside Plot
ax = fig.add_axes((0.8, 0.8, 0.2, 0.2)) ## Either [] or () is fine
ax.plot(x,y[:,2], color = 'r')
ax.set_xticks([])
```

Scatter

```python
np.random.seed(10)

x = np.random.randint(1, 10,size=100)
y = np.random.randint(1, 10,size=100)

sizes = np.random.uniform(5, 200, len(x))
colors = np.random.uniform(0, 100, len(x))

fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100, alpha=0.7)
ax.set(xlim=(0, 10), xticks=np.arange(1, 10),
       ylim=(0, 10), yticks=np.arange(1, 10))

plt.show()
```

```python
np.random.seed(15)
num_data = 80

x = np.array(range(num_data))
y = np.random.randn(num_data)

sizes = 50 * np.random.randn(num_data)**2
colors = np.random.randn(num_data)

fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, alpha=0.7)


plt.show()
```

```python
np.random.seed(9989)
num_points = 100
# gradient = 0.5
x = np.array(range(num_points))
y = np.random.randn(num_points) * 10 + x
fig, ax = plt.subplots(figsize=(8, 4))
colors = np.random.rand(num_points)
size = np.pi * (2 + np.random.rand(num_points) * 8) ** 2
ax.scatter(x, y, s=size, c=colors, alpha=0.5)

fig.suptitle('A Simple Scatterplot')
```

```python

np.random.seed(36)
num_data = 100

x = np.array(range(num_data))
print(x)
y = np.random.randn(num_data) * 10 + x
print(y)

fig, ax = plt.subplots(figsize=(12, 6))
fig.suptitle('Linear Regression')

ax.scatter(x, y)
m, c = np.polyfit(x, y, 1) #linear regression, deg = 1
z = m*x + c #regression line
ax.plot(x, z)
```
